```python

Women = [115, 215, 250, 200,600]
Men = [114, 230, 510, 370,700]

n= 5
r = np.arange(n)
width = 0.25


plt.bar(r, Women, color = 'b',
		width = width, edgecolor = 'black',
		label='Women')
plt.bar(r + width, Men, color = 'g',
		width = width, edgecolor = 'black',
		label='Men')

plt.xlabel("Year")
plt.ylabel("Number of people voted")
plt.title("Number of people voted in each year")

# plt.grid(linestyle='--')
plt.xticks(r+width,['2018','2019','2020','2021','2022'])
plt.legend(title = "Gender ")

plt.show()
```

```python
# visualization with seaborn (wkm-juex-cqi)
import seaborn as sns

data = sns.load_dataset('iris')

data.head()
```

```python
# fig size in seaborn
sns.lineplot(x='sepal_length',y = 'sepal_width',data = data)
plt.title('My Graph')
sns.despine()
plt.figure(figsize = (10,8))
```

```python
# using despine in seaborn
sns.lineplot(x='sepal_length',y = 'sepal_width',data = data)
plt.title('My Graph')
sns.despine()
```

```python
# style on seaborn
sns.lineplot(x='sepal_length',y = 'sepal_width',data = data)
plt.title('My Graph')
sns.set_style("dark")
```

```python
# subplot in sns
def plot():
  sns.lineplot(x='sepal_length',y = 'sepal_width',data = data)
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    plot()

plt.subplot(212)
plot()
```
